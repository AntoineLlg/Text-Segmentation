```python
import numpy as np
import matplotlib.pyplot as plt
import PIL
from skimage import io as skio
import os
from utils import rotation
from sklearn.linear_model import LinearRegression
```

```python
def view_image(im, figsize=(10, 10)):
    fig = plt.figure(figsize=figsize)
    plt.imshow(im, 'Greys_r')
    plt.axis('off')
    plt.show()
```

```python
image = PIL.Image.open("data/image2.png").convert('L') # convert to greyscale
image = image.rotate(13, resample=PIL.Image.BICUBIC)
A, B = image.size
N = A//2
M = B//2
square = image.crop((N-200, M-200, N+200, M+200))
square
```

```python
im = np.array(square)
```

```python
view_image(im)
```

```python
IM = np.fft.fftshift(np.fft.fft2(im))
view_image(np.log(np.abs(IM)))
```

```python
N, M = im.shape
im[:N//2, :M//2].shape
```

```python
def quadrants(im):
    N, M = im.shape
    yield im[:N//2, :M//2]
    yield im[:N//2, M//2:]
    yield im[N//2:, :M//2]
    yield im[N//2:, M//2:]
    
def middle_square(shape, n, m=None):
    """
    creates a matrix of ones excepts for a rectangle of size 2n * 2m in the middle overlapping equally each quadrant
    
    :parameters:
    n int: height of the rectangle. If m is not used, n is also the width 
    m int (optional): width of the rectangle
    """
    if not m:
        m = n
    N, M = shape
    res = np.ones(shape)*255
    res[N//2-n:N//2+n, M//2-m:M//2+m] = 0
    return res
```

```python
view_image(middle_square(IM.shape, 10))
```

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 10))
ax = ax.flatten()
for quad, axis in zip(quadrants(np.log(np.abs(IM)+1)*middle_square(IM.shape, 5)), ax):
    axis.imshow(quad, 'Greys_r')
    axis.axis('off')
```

```python
def find_brightest_pixels(image, N):
    image_1d = image.flatten()
    idx_1d = image_1d.argsort()[-N:]
    x_idx, y_idx = np.unravel_index(idx_1d, image.shape)
    return x_idx, y_idx
```

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 10))
ax = ax.flatten()
for quad, axis in zip(quadrants(np.log(np.abs(IM)+1)*middle_square(IM.shape, 5)), ax):
    test = np.zeros(quad.shape)
    for x, y in zip(*find_brightest_pixels(quad, 20)):
        test[x, y] = 1
    axis.imshow(test, 'Greys_r')
    axis.axis('off')
```

```python
coef = []

for quad in quadrants(np.log(np.abs(IM) + 1)*middle_square(IM.shape, 20)):
    X, y = find_brightest_pixels(quad, 30)
    reg = LinearRegression().fit(X.reshape(-1, 1), y)
    R2 = reg.score(X.reshape(-1, 1), y)
    if R2>0.9:
        coef.append(reg.coef_)
if len(coef) == 0:
    for quad in quadrants(np.log(np.abs(IM) + 1)*middle_square(IM.shape, 20)):
        X, y = find_brightest_pixels(quad, 30)
        reg = LinearRegression().fit(X.reshape(-1, 1), y)
        R2 = reg.score(X.reshape(-1, 1), y)
        coef.append(reg.coef_)
        
rotated = rotation(im, np.arctan(np.mean(coef)), clip=False)
rotated*= (rotated>50)
view_image(rotated)
```

```python
plt.plot(rotated.sum(axis=-1))
plt.plot([rotated.sum(axis=-1).max()*0.95]*rotated.shape[0])
plt.show()
```

```python
def cut_lines(im):
    white_density = im.sum(axis=-1)
    threshold = .95 * white_density.max()
    
    text_line = white_density < threshold
    init = False
    for i in range(len(white_density)):
        if init:
            if text_line[i]:
                end = i
            else:
                init = False
                yield start, end + 1
        else:
            if text_line[i]:
                start = i
                end = i
                init = True
```

```python
for u, v in cut_lines(rotated):
    fig = plt.figure(figsize=(20,1))
    plt.imshow(rotated[u:v, :], 'Greys_r')
    plt.show()
```

```python
u, v = next(iter(cut_lines(rotated)))
```

```python
def cut_words(line):
    white_density = line.sum(axis=0)
    threshold = .98 * white_density.max()
    
    text_line = white_density < threshold
    init = False
    for i in range(len(white_density)):
        if init:
            if text_line[i]:
                end = i
            else:
                init = False
                yield start, end + 1
        else:
            if text_line[i]:
                start = i
                end = i
                init = True
```

```python
for p, q  in cut_words(rotated[u:v]):
    fig = plt.figure(figsize=(10,1))
    plt.imshow(rotated[u:v, p:q], 'Greys_r')
    plt.show()
```

```python
from PIL import ImageDraw

def colors_generator():
    color_bank = [(251 , 180 , 174 , 140),
                  (179 , 205 , 227 , 140),
                  (204 , 235 , 197 , 140),
                  (222 , 203 , 228 , 140),
                  (254 , 217 , 166 , 140),
                  (255 , 255 , 204 , 140),
                  (229 , 216 , 189 , 140),
                  (253 , 218 , 236 , 140)]
    n = 0
    while True:
        yield color_bank[n % 8]
        n += 1


def segment(image):
    res = PIL.Image.fromarray(image).convert('RGBA')
    mask = PIL.Image.new('RGBA', res.size, color=(255, 255, 255, 0))
    draw = ImageDraw.Draw(mask)
    colors = colors_generator()
    for u, v in cut_lines(image):
        for k, l in cut_words(image[u:v]):
            draw.rectangle(((k, u), (l, v)), fill=next(iter(colors)))
            
    res.alpha_composite(mask)
    return res

segment(rotated)
```
